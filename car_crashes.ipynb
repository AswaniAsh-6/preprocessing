```python
iqr=q3-q1
print(iqr)
```

```python
lower=q1-1.5*iqr
upper=q3+1.5*iqr
print(lower)
print(upper)
```

```python
outlier=[]
for i in data['alcohol']:
    if((i>upper) or (i<lower)):
        outlier.append(i)
```
